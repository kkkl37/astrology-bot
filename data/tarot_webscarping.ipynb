```python
from os.path import exists
from bs4 import BeautifulSoup
import re
import requests
import time
import pandas as pd
```

```python
tarot_cards = {
    'major-arcana': ['fool', 'magician', 'high priestess', 'empress', 'emperor', 'hierophant', 'lovers', 
                     'chariot', 'strength', 'hermit', 'wheel of fortune', 'justice', 'hanged-man', 'death',
                     'temperance', 'devil', 'tower', 'star', 'moon', 'sun', 'judgement', 'world'],
    'minor-arcana': {'suit-of-cups': ['ace-of-cups', 'two-of-cups', 'three-of-cups', 'four-of-cups', 'five-of-cups',
                                        'six-of-cups', 'seven-of-cups', 'eight-of-cups', 'nine-of-cups', 'ten-of-cups',
                                        'page-of-cups', 'knight-of-cups', 'queen-of-cups', 'king-of-cups'],
                        'suit-of-pentacles': ['ace-of-pentacles', 'two-of-pentacles', 'three-of-pentacles', 'four-of-pentacles',
                                             'five-of-pentacles', 'six-of-pentacles', 'seven-of-pentacles', 'eight-of-pentacles',
                                             'nine-of-pentacles', 'ten-of-pentacles', 'page-of-pentacles', 'knight-of-pentacles',
                                             'queen-of-pentacles', 'king-of-pentacles'],
                        'suit-of-swords': ['ace-of-swords', 'two-of-swords', 'three-of-swords', 'four-of-swords', 'five-of-swords',
                                         'six-of-swords', 'seven-of-swords', 'eight-of-swords', 'nine-of-swords', 'ten-of-swords',
                                         'page-of-swords', 'knight-of-swords', 'queen-of-swords', 'king-of-swords'],
                        'suit-of-wands': ['ace-of-wands', 'two-of-wands', 'three-of-wands', 'four-of-wands', 'five-of-wands',
                                        'six-of-wands', 'seven-of-wands', 'eight-of-wands', 'nine-of-wands', 'ten-of-wands',
                                        'page-of-wands', 'knight-of-wands', 'queen-of-wands', 'king-of-wands']}

    }
```

```python
# scrape the horoscope from the website and save them to a csv file with two columns: horoscope and text, ignore the date
def scrape_tarot(path):
    if not exists(path):
        with open(path, 'a') as f:
            f.write(f'kind,card,upright_keywords,reverse_keywords,description,upright,reversed\n')
    for kind in tarot_cards.keys():
        if kind == 'major-arcana':
            for card in tarot_cards[kind]:
                url = f'https://biddytarot.com/tarot-card-meanings/{kind}/{card}/'
                
                response = requests.get(url)
                soup = BeautifulSoup(response.text, 'html.parser')
                keywords_description = soup.find('div', class_='col span_8 order-md-1')
                for i, childp in enumerate(keywords_description.find_all('p')):
                    if i == 0:
                        upright_keywords = childp.text.strip().split(": ")[1]
                    elif i == 1:
                        reverse_keywords = childp.text.strip().split(": ")[1]
                    elif i == 2:
                        description = childp.text.strip()
                    elif i > 2 and i != len(keywords_description.find_all('p')) - 1:
                        description = description+' '+childp.text.strip()
                
                uprights = soup.find('div', attrs={"class": "container", "id": "upright"}).find_all('p')
                upright = ''
                for text in uprights:
                    upright = upright + text.text.strip() + ' '
                
                reverseds = soup.find('div', attrs={"class": "container", "id": "reversed"}).find_all('p')
                reversed = ''
                for text in reverseds:
                    reversed = reversed + text.text.strip() + ' '
                with open(path, 'a') as f:
                    f.write(f'{kind},{card},"{upright_keywords}","{reverse_keywords}","{description}","{upright}","{reversed}"\n')
                    print(f'{kind},{card} done')
        else:
            for suit in tarot_cards[kind].keys():
                for card in tarot_cards[kind][suit]:
                    url = f'https://biddytarot.com/tarot-card-meanings/{kind}/{suit}/{card}/'
        
                    response = requests.get(url)
                    soup = BeautifulSoup(response.text, 'html.parser')
                    keywords_description = soup.find('div', class_='col span_8 order-md-1')
                    for i, childp in enumerate(keywords_description.find_all('p')):
                        if i == 0:
                            upright_keywords = childp.text.strip().split(": ")[1]
                        elif i == 1:
                            reverse_keywords = childp.text.strip().split(": ")[1]
                        elif i == 2:
                            description = childp.text.strip()
                        elif i > 2 and i != len(keywords_description.find_all('p')) - 1:
                            description = description+' '+childp.text.strip()

                    upright, reversed = '', ''

                    details = soup.find_all('div', attrs={"class": "container"})
                    for i, detail in enumerate(details):
                        if detail.find('div', attrs={"class": "col span_12"}) and i>2:
                            uprights = detail.find_all('p')
                            for text in uprights:
                                upright = upright + text.text.strip() + ' '

                        if detail.find('div', attrs={"class": "col span_8"}):
                            reverseds = detail.find_all('p')
                            for text in reverseds:
                                reversed = reversed + text.text.strip() + ' '

                    with open(path, 'a') as f:
                        f.write(f'{kind},{card},"{upright_keywords}","{reverse_keywords}","{description}","{upright}","{reversed}"\n')
                    print(f'{kind},{card} done')

        time.sleep(1)
```

```python
path = 'data/tarot.csv'
scrape_tarot(path)
```
