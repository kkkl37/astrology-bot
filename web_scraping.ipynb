```python
from os.path import exists
from bs4 import BeautifulSoup
import re
import requests
import time
import pandas as pd
```

```python
zodiacs = ['aries', 'taurus', 'gemini', 'cancer', 'leo', 'virgo', 'libra', 'scorpio', 'sagittarius', 'capricorn', 'aquarius', 'pisces']
weekday = ['monday', 'tuesday', 'wednesday', 'thursday', 'friday', 'saturday', 'sunday']
subarea = ['daily','love', 'work', 'dating']
```

```python
# scrape the horoscope from the website and save them to a csv file with two columns: horoscope and text, ignore the date
def scrape_horoscope():
    if not exists('horoscope.csv'):
        with open('horoscope.csv', 'a') as f:
            f.write(f'zodiac,area,text\n')
    for zodiac in zodiacs:
        for area in subarea:
            if area == 'daily':
                url = f'https://www.astrology.com/horoscope/daily/{zodiac}.html'
            else:
                url = f'https://www.astrology.com/horoscope/daily-{area}/{zodiac}.html'
            response = requests.get(url)
            soup = BeautifulSoup(response.text, 'html.parser')
            horoscope = soup.find('div', class_='horoscope-content-wrapper')
            text = horoscope.find(attrs={'id':'content'}).text
            with open('horoscope.csv', 'a') as f:
                f.write(f'{zodiac},{area},"{text}"\n')
            print(f'{zodiac} {area} done')
            time.sleep(1)
```

```python
# start the program
# if not exists('horoscope.csv'):
#     scrape_horoscope()
scrape_horoscope()
```
